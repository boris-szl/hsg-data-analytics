```python
# Code below
def median(series):
    array = series.to_numpy()
    array = sorted(array)
    n = len(array) # series length
    if (n % 2 == 0):
         # if lenght is even the median is the average of the sum of x[n+1] andbecause index starts at 0 is 
        median = 1/2*(array[int(n/2)] + array[int(n/2+1)])
    else:
        median = array[int(n+1/2)]
    return median
```
